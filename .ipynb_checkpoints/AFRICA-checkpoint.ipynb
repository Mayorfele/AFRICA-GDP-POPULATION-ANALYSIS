REGIONAL ANALYSIS

Regional summary

```python
def get_regional_summary_by_year(data, year):
    # Filter the data for the given year
    africa_year = africa[africa['Year'] == year]
    
    # Check if there is any data for the given year
    if africa_year.empty:
        print(f"No data available for the year {year}.")
        return None
    
    # Perform the aggregation
    regional_summary = africa_year.groupby('Continent').agg({
        'Population ': 'sum',
        'GDP (USD)': 'sum',
        'GDP per Capita': 'mean'
    }).reset_index()
    
    return regional_summary

# Example usage:
year = 2021  # Replace with the desired year
regional_summary = get_regional_summary_by_year(africa, year)

regional_summary
```

```python
# Plot Pie Charts
plt.figure(figsize=(18, 8))

# Population Pie Chart
plt.subplot(1, 3, 1)
plt.pie(regional_summary['Population '], labels=regional_summary['Continent'], autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Continent', fontweight='bold')

# GDP Pie Chart
plt.subplot(1, 3, 2)
plt.pie(regional_summary['GDP (USD)'], labels=regional_summary['Continent'], autopct='%1.1f%%', startangle=140)
plt.title('GDP Distribution by Continent', fontweight='bold')

# GDP per Capita Pie Chart
plt.subplot(1, 3, 3)
plt.pie(regional_summary['GDP per Capita'], labels=regional_summary['Continent'], autopct='%1.1f%%', startangle=140)
plt.title('GDP per Capita Distribution by Continent',  fontweight='bold')

plt.tight_layout()


# Save the image
plt.savefig('Africa pie.png', bbox_inches='tight')

plt.show()
```

```python
def draw_choropleth_map(data, year, width=800, height=400):
    # Filter data for the specified year
    data_year = data[data['Year'] == year]
    
    # Check if there is any data for the specified year
    if data_year.empty:
        print(f"No data available for the year {year}.")
        return
    
    # Create choropleth map
    fig = px.choropleth(
        data_year,
        locations="Country",
        locationmode='country names',
        color="GDP per Capita",
        hover_name="Country",
        color_continuous_scale=px.colors.sequential.Plasma,
        title=f"GDP per Capita in Africa for {year}",
    )
    
    # Update layout to set figure size and margins
    fig.update_layout(
        width=width,
        height=height    
    )
    
    fig.update_geos(scope="africa")
    fig.show()

# Example usage:
year = 2022  # Replace with the desired year
draw_choropleth_map(africa, year, width=800, height=400)
```

```python
def rank_countries_by_gdp(data, year, continent):
    # Filter the data for the specified year and continent
    filtered_data = africa[(africa['Year'] == year) & (africa['Continent'] == continent)]
    
    # Check if there is any data for the specified year and continent
    if filtered_data.empty:
        print(f"No data available for the year {year} and continent {continent}.")
        return
    
    # Rank the countries by GDP
    ranked_data = filtered_data.sort_values(by='GDP (USD)', ascending=False)
    
    # Plot the bar chart
    plt.figure(figsize=(12, 8))
    sns.barplot(x='GDP (USD)', y='Country', data=ranked_data, palette='deep')
    plt.title(f'Ranking Countries in {continent} by GDP for the Year {year}',  fontweight='bold')
    plt.xlabel('GDP (USD)')
    plt.ylabel('Country')
    
    # Save the image
    plt.savefig('Africa rank ef.png', bbox_inches='tight')
    
    plt.show()

# Example usage:
year = 2014  # Replace with the desired year
continent = 'East Africa'  # Replace with the desired continent
rank_countries_by_gdp(africa, year, continent)
```

Population Growth

```python
def population_trend(country_name):
    # Filter the data for the specified country
    country_data = africa[africa['Country'] == country_name]
    
    # Check if there is any data for the specified country
    if country_data.empty:
        print(f"No data available for the country {country_name}.")
        return
    
    # Plot the GDP per Capita trend over the years
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=country_data, x='Year', y='Population ', marker='o')
    plt.title(f'Population Trend For {country_name}')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.grid(True)
    
    plt.savefig('south africa rank POP.png', bbox_inches='tight')
    
    plt.show()

# Example usage:
country_name = 'South Africa'
population_trend(country_name)
```

CORRELATION

```python
def analyze_correlation(africa, country_name):
    # Filter the data for the specified country
    country_data = africa[africa['Country'] == country_name]
    
    # Check if there is any data for the specified country
    if country_data.empty:
        print(f"No data available for the country {country_name}.")
        return
    
    # Calculate the Pearson correlation coefficient
    correlation = country_data['GDP per Capita'].corr(country_data['Population '])
    print(f"Pearson correlation coefficient between GDP and Population for {country_name}: {correlation:.2f}")
    
    # Create a scatter plot with a regression line
    plt.figure(figsize=(10, 6))
    sns.regplot(data=country_data, x='Population ', y='GDP per Capita', scatter_kws={'s':50}, line_kws={'color':'blue'})
    plt.title(f'GDP vs Population for {country_name}')
    plt.xlabel('Population')
    plt.ylabel('GDP (USD)')
    plt.grid(True)
    
    plt.savefig('GH reg ef.png', bbox_inches='tight')
    
    plt.show()

# Example usage:
country_name = 'Ghana'  # Replace with the desired country name
analyze_correlation(africa, country_name)
```

```python
def plot_correlation_heatmap(data, country_name):
    # Filter the data for the specified country
    country_data = data[data['Country'] == country_name]
    
    # Check if there is any data for the specified country
    if country_data.empty:
        print(f"No data available for the country {country_name}.")
        return
    
    # Calculate the correlation matrix
    correlation_matrix = country_data[['Population ', 'GDP (USD)']].corr()
    
    # Create a heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Correlation Heatmap for {country_name}')
    plt.show()

# Example usage:
country_name = 'Nigeria'  # Replace with the desired country name
plot_correlation_heatmap(africa, country_name)
```
